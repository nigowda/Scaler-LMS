### **Group based Apply**

```python
# I want to find out all the risky movies
```

```python
def calculate_risky(x):
    x["risky"]=(x["budget_mil"]-x["revenue_mil"].mean())>=0
    return x
```

```python
df_risky=df.groupby(["director_name"]).apply(calculate_risky)
df_risky
```

```python
df_risky["risky"]==True
```

```python
df_risky.loc[df_risky["risky"]==True]
```
