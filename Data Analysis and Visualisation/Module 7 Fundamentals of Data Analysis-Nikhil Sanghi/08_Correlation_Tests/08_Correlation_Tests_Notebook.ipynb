```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
```

```python
# Lalit
heights = np.array([68, 62, 64, 61, 70, 66, 61, 65, 71, 72]) # inches
weights = np.array([72, 58, 67, 72, 79, 61, 68, 64, 80, 79]) # kgs

sns.scatterplot(x=heights, y=weights)
plt.axvline(heights.mean(),color="green",linestyle="--")
plt.axhline(weights.mean(),color="green",linestyle="--")

plt.xlabel("Heights (inches)")
plt.ylabel("Weights (Kgs)")
plt.show()
```

```python
covariance_lalit=((heights-heights.mean())*(weights-weights.mean())).sum()/len(heights)
covariance_lalit
```

```python
correlation_lalit= covariance_lalit/(heights.std()*weights.std())
correlation_lalit
```

```python
# Nakul
heights1 = np.array([68, 62, 64, 61, 70, 66, 61, 65, 71, 72])*2.54 # cms
weights1 = np.array([72, 58, 67, 72, 79, 61, 68, 64, 80, 79])*2.2 # lbs

sns.scatterplot(x=heights1, y=weights1)
plt.axvline(heights1.mean(),color="green",linestyle="--")
plt.axhline(weights1.mean(),color="green",linestyle="--")

plt.xlabel("Heights (cms)")
plt.ylabel("Weights (lbs)")
plt.show()
```

```python
covariance_nakul=((heights1-heights1.mean())*(weights1-weights1.mean())).sum()/len(heights1)
covariance_nakul
```

```python
correlation_nakul= covariance_nakul/(heights1.std()*weights1.std())
correlation_nakul
```

```python
# Darsh
heights2 = np.array([68, 62, 64, 61, 70, 66, 61, 65, 71, 72])*2.54 # cms
weights2 = np.array([72, 58, 67, 72, 79, 61, 68, 64, 80, 79])*1000 # gms

sns.scatterplot(x=heights2, y=weights2)
plt.axvline(heights2.mean(),color="green",linestyle="--")
plt.axhline(weights2.mean(),color="green",linestyle="--")

plt.xlabel("Heights (cms)")
plt.ylabel("Weights (gms)")
plt.show()
```

```python
covariance_darsh=((heights2-heights2.mean())*(weights2-weights2.mean())).sum()/len(heights2)
covariance_darsh
```

```python
correlation_darsh= covariance_darsh/(heights2.std()*weights2.std())
correlation_darsh
```

```python
df=pd.DataFrame([heights,weights]).transpose()
df.rename(columns={0:"Heights",1:"Weights"},inplace=True)
df
```

```python
df_rank=df.rank()
df_rank.rename(columns={"Heights":"Heights_rank","Weights":"Weights_rank"},inplace=True)
df_rank
```

```python
pearsonr(df_rank["Heights_rank"],df_rank["Weights_rank"])
```
